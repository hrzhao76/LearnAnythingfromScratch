```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_s = np.linspace(-5, 10, 100)

y1_s = -3/4 * x_s - 2
y2_s = -4/3 * x_s - 7/3


plt.plot(x_s, y1_s, label='y1')
plt.plot(x_s, y2_s, label='y2')
```

```python
from scipy.stats import pearsonr
pearsonr(x_s + x_s, y1_s + y2_s)
```

```python
from scipy.optimize import fsolve
def func(xs):
    x, y = xs
    return [3 * x + 4 * y + 8,
            4 * x + 3 * y + 7]
```

```python
root = fsolve(func, [1, 1])
```

```python
y1, y2 = root
```

```python
y2
```

```python
pearsonr([y1], [y2])
```
